```python
import random
%matplotlib inline
import matplotlib.pyplot as plt
def func(r , x):
    return r*x*(1-x)
def plot(lambdas ,xs):
    plt.figure(figsize=(10,10))
    for i, r in enumerate(lambdas):
        plt.scatter([r] * len(xs[i]), xs[i], s=0.1, color='black')
    plt.xlabel("lambda")
    plt.ylabel("x")
    plt.show()
if __name__ == '__main__':
    random.seed(1)
    x0 = random.random()
    N = 10000
    lambdas = [0.01 * n for n in range(401)]
    xs = []
    for r in lambdas:
        temp = []
        x = x0
        for _ in range(N//2):  
            x = func(r, x)
        
        for _ in range(N//2, N): 
            x = func(r, x)
            temp.append(x)
        
        xs.append(temp)
    plot(lambdas,xs)
```
