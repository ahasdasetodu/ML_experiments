Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji $f(x)=x^2$, startując z punktu $x=2$. Zilustruj całą sytuacje oznaczając na wykresie 5 pierwszych iteracji.

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2

x = np.linspace(-5, 5, 200)
y = f(x)
plt.plot(x, y, '--k', ms=10)


def step_gradient_1d(x_current, learningRate):
    x_gradient = 2*x_current
    new_x = x_current - learningRate*x_gradient
    
    plt.arrow(x_current, f(x_current), - (learningRate * x_gradient), -(f(x_current)-f(new_x)),
              head_width=0.05, head_length=0.5,ec="red")
        
    return new_x
def gradient_descent_runner_1d(starting_x, learning_rate, num_iterations):
    x = starting_x
    print(x)
    for i in range(num_iterations):
        x = step_gradient_1d(x,learning_rate)
        #print(x)
    return x

learning_rate = 0.2
initial_x = 5 
num_iterations = 30
x = gradient_descent_runner_1d(initial_x, learning_rate, num_iterations)

plt.show()
```

# Zadanie 2

Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji 

$$
f(x,y)=4*x^2-2x+y^2,
$$ 

startując z punktu $(x,y)=(2,2)$. Zilustruj całą sytuacje oznaczając na wykresie kilka pierwszych iteracji.
Wykonaj obrazek 3D.

```python
###########

import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
#%matplotlib notebook

plt.close('all')

fun = lambda x,y: 4*x**2+y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Make data.
X = np.arange(-7, 7, 0.25)
Y = np.arange(-7, 7, 0.25)
X, Y = np.meshgrid(X, Y)
Z = fun(X, Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0.01, antialiased=True, alpha=0.3)


#####################################

def step_gradient_2d(x_current, y_current, learningRate):
    x_gradient = 8*x_current-2
    y_gradient = 2*y_current
    
    new_x = x_current - learningRate*x_gradient
    new_y = y_current - learningRate*y_gradient
    
    ax.quiver(x_current, y_current, (fun(x_current, y_current)) ,
              - (learningRate * x_gradient), - (learningRate * y_gradient), 
              (-(fun(x_current,y_current)-fun(new_x,new_y)))) 
    
    return [new_x, new_y]

def gradient_descent_runner_2d(starting_x, starting_y, learning_rate, num_iterations):
    x = starting_x
    y = starting_y
    for i in range(num_iterations):
        x, y = step_gradient_2d(x, y, learning_rate)
        #print(x, y)
    return [x, y]


learning_rate = 0.2
initial_x = 0 # initial y-intercept guess
initial_y = 5 # initial slope guess
num_iterations = 10
[x, y] = gradient_descent_runner_2d(initial_x, initial_y, learning_rate, num_iterations)


#####################################

plt.plot([initial_x],[initial_y],[fun(initial_x,initial_y)],"ok")
plt.show()
```

Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji $f(x)=x^2-y^2$, startując z punktu $x=5$ oraz $x=1$. Zilustruj całą sytuacje oznaczając na wykresie 5 pierwszych iteracji.

Wykonaj rysunek:
  
  * 2D conture plot
  * 3D 
  
Co się stanie gdy zaczniemy z punktu $x=5$ oraz $x=0$? 

```python
#my solution
def f(x,y):
    return x**2 - y**2
def grad_f(x,y):
    return 2*x, -2*y

def gradient_descent(x,y, learning_rate):
    x_grad, y_grad = grad_f(x,y)
    x -= learning_rate*x_grad
    y -= learning_rate*y_grad
    return x,y
n_epochs = 20

x = 5
y = 1
steps1 = [(x,y)]
learning_rate = 0.05
for epoch in range(n_epochs):
    x, y = gradient_descent(x, y, learning_rate)
    steps1.append((x,y))
steps1 = np.array(steps1)
x = 5
y = 0
steps2 = [(x,y)]
learning_rate = 0.05
for epoch in range(n_epochs):
    x, y = gradient_descent(x, y, learning_rate)
    steps2.append((x,y))
steps2 = np.array(steps2)

xs = np.linspace(-5, 5, 200)
ys = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(xs, ys)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
x_steps1 = np.array(steps1[:,0])
y_steps1 = np.array(steps1[:,1])
ax.plot(x_steps1, y_steps1,f(x_steps1,y_steps1),'r')
x_steps2 = np.array(steps2[:,0])
y_steps2 = np.array(steps2[:,1])
ax.plot(x_steps2, y_steps2,f(x_steps2,y_steps2),'b')
plt.show()

plt.contour(X, Y, Z)
plt.plot(x_steps1, y_steps1)
plt.plot(x_steps2, y_steps2)
plt.show()
```

```python
#chatgpt solution 
# Function f(x, y) = x^2 - y^2
def f(x, y):
    return x**2 - y**2

# Gradient of f(x, y)
def gradient(x, y):
    df_dx = 2*x
    df_dy = -2*y
    return np.array([df_dx, df_dy])

# Gradient Descent Algorithm
def gradient_descent(start_x, start_y, learning_rate, iterations):
    x, y = start_x, start_y
    path = [(x, y)]  # Store points for visualization
    
    for _ in range(iterations):
        grad = gradient(x, y)  # Compute gradient
        x -= learning_rate * grad[0]  # Update x
        y -= learning_rate * grad[1]  # Update y
        path.append((x, y))  # Store new point
    
    return np.array(path)

# Run Gradient Descent
learning_rate = 0.2
iterations = 5
start_points = [(5, 1), (5, 0)]  # Two cases: (5,1) and (5,0)

# Generate Meshgrid for plotting
X = np.linspace(-6, 6, 100)
Y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

# 2D Contour Plot
plt.figure(figsize=(10, 5))
plt.contour(X, Y, Z, levels=30, cmap="coolwarm")
plt.colorbar(label="f(x, y)")

for start_x, start_y in start_points:
    path = gradient_descent(start_x, start_y, learning_rate, iterations)
    plt.plot(path[:, 0], path[:, 1], 'o-', label=f"Start: ({start_x}, {start_y})")

plt.title("Gradient Descent Trajectory on Contour Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# 3D Surface Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="coolwarm", alpha=0.6)

for start_x, start_y in start_points:
    path = gradient_descent(start_x, start_y, learning_rate, iterations)
    ax.plot(path[:, 0], path[:, 1], f(path[:, 0], path[:, 1]), 'o-', label=f"Start: ({start_x}, {start_y})")

ax.set_title("Gradient Descent Trajectory on 3D Surface")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.legend()
plt.show()
```
