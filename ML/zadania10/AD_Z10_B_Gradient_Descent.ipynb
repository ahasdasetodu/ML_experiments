Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji $f(x)=x^2$, startując z punktu $x=2$. Zilustruj całą sytuacje oznaczając na wykresie 5 pierwszych iteracji.

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2

x = np.linspace(-5, 5, 200)
y = f(x)
plt.plot(x, y, '--k', ms=10);


def step_gradient_1d(x_current, learningRate):
    x_gradient = 2*x_current
    new_x = x_current - learningRate * x_gradient
    
    plt.arrow(x_current, f(x_current), - (learningRate * x_gradient), -(f(x_current)-f(new_x)),
              head_width=0.05, head_length=0.5,ec="red")
        
    return new_x
def gradient_descent_runner_1d(starting_x, learning_rate, num_iterations):
    x = starting_x
    print(x)
    for i in range(num_iterations):
        x = step_gradient_1d(x,learning_rate)
        #print(x)
    return x

learning_rate = 0.2
initial_x = 5 
num_iterations = 30
gradient_descent_runner_1d(initial_x, learning_rate, num_iterations)

plt.show()
```

# Zadanie 2

Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji 

$$
f(x,y)=4*x^2-2x+y^2,
$$ 

startując z punktu $(x,y)=(2,2)$. Zilustruj całą sytuacje oznaczając na wykresie kilka pierwszych iteracji.
Wykonaj obrazek 3D.

```python
###########

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

%matplotlib inline

plt.close('all')

fun = lambda x,y: 4*x**2+y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Make data.
X = np.arange(-7, 7, 0.25)
Y = np.arange(-7, 7, 0.25)
X, Y = np.meshgrid(X, Y)
Z = fun(X, Y)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0.01, antialiased=True, alpha=0.3)


#####################################

def step_gradient_2d(x_current, y_current, learningRate):
    x_gradient = 8 * x_current - 2
    y_gradient = 2*y_current
    
    new_x = x_current - learningRate * x_gradient
    new_y = y_current - learningRate * y_gradient
    

    return [new_x, new_y]

def gradient_descent_runner_2d(starting_x, starting_y, learning_rate, num_iterations):
    x = starting_x
    y = starting_y
    x_steps = [x]
    y_steps = [y]
    z_steps = [fun(x,y)]
    for i in range(num_iterations):
        x, y = step_gradient_2d(x, y, learning_rate)
        x_steps.append(x)
        y_steps.append(y)
        z_steps.append(fun(x,y))
        #print(x, y)
    ax.plot(x_steps, y_steps, z_steps, color='red', linewidth=2, label='Trajectory')
    return [x, y]


learning_rate = 0.15
initial_x = 2 # initial y-intercept guess
initial_y = 2 # initial slope guess
num_iterations = 10
[x, y] = gradient_descent_runner_2d(initial_x, initial_y, learning_rate, num_iterations)


#####################################

plt.plot([initial_x],[initial_y],[fun(initial_x,initial_y)],"ok")
plt.show()
```

Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji $f(x)=x^2-y^2$, startując z punktu $x=5$ oraz $x=1$. Zilustruj całą sytuacje oznaczając na wykresie 5 pierwszych iteracji.

Wykonaj rysunek:
  
  * 2D conture plot
  * 3D 
  
Co się stanie gdy zaczniemy z punktu $x=5$ oraz $x=0$? 

```python

#my solution
f = lambda x,y: x**2 - y**2



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.arange(-7, 7, 0.25)
Y = np.arange(-7, 7, 0.25)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0.01, antialiased=True, alpha=0.3)



def gradient_step(curr_x,curr_y,learning_rate):
    gradient_x = 2 * curr_x
    gradient_y = -2 * curr_y
    new_x = curr_x - learning_rate * gradient_x
    new_y = curr_y - learning_rate * gradient_y
    return [new_x, new_y]
    
def find_minimum(starting_x, starting_y, learning_rate, num_iterations):
    x_steps = [starting_x]
    y_steps = [starting_y]
    z_steps = [f(starting_x,starting_y)]
    for i in range(num_iterations):
        x,y = gradient_step(x_steps[-1],y_steps[-1],learning_rate)
        x_steps.append(x)
        y_steps.append(y)
        z_steps.append(f(x,y))

    return [x_steps, y_steps, z_steps]


starting_x = 5
starting_y = 0
learning_rate = 0.05
plot_steps_amt = 5
x,y,z = find_minimum(starting_x, starting_y, learning_rate, num_iterations)
print(x[-1], y[-1], z[-1])
ax.plot(x[:plot_steps_amt], y[:plot_steps_amt], z[:plot_steps_amt], color='red', linewidth=2, label='Trajectory')
plt.show()

fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('x')
ax.set_ylabel('y')
CS = ax.contour(X,Y,Z)
ax.plot(x[:plot_steps_amt], y[:plot_steps_amt],  color='red', linewidth=2, label='Trajectory')
plt.show()



#chatgpt solution
# Funkcja i jej gradient
def f(x, y):
    return x**2 - y**2

def grad_f(x, y):
    return np.array([2*x, -2*y])

# Gradient descent
def gradient_descent(starting_point, learning_rate=0.1, steps=5):
    point = np.array(starting_point)
    trajectory = [point.copy()]
    
    for _ in range(steps):
        grad = grad_f(point[0], point[1])
        point = point - learning_rate * grad
        trajectory.append(point.copy())
    
    return np.array(trajectory)

# Trajektorie dla dwóch punktów startowych
traj1 = gradient_descent([5, 1])
traj2 = gradient_descent([5, 0])  # <== przypadek do analizy

# Siatka do rysowania powierzchni
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# === WYKRES 2D: contour plot z trajektorią ===
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=30, cmap='coolwarm')
plt.clabel(contour)
plt.plot(traj1[:, 0], traj1[:, 1], 'o-r', label='start: (5,1)')
plt.plot(traj2[:, 0], traj2[:, 1], 'o-g', label='start: (5,0)')
plt.title('Gradient Descent – 2D Contour Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.axis('equal')
plt.show()

# === WYKRES 3D ===
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.6)

# Trajektorie na 3D
ax.plot(traj1[:, 0], traj1[:, 1], f(traj1[:, 0], traj1[:, 1]), 'o-r', label='start: (5,1)')
ax.plot(traj2[:, 0], traj2[:, 1], f(traj2[:, 0], traj2[:, 1]), 'o-g', label='start: (5,0)')

ax.set_title("Gradient Descent – 3D Surface")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.legend()
plt.show()
```
