````python
#low noise, plenty of samples, should be easy
X, y = sklearn.datasets.make_moons(n_samples=1000, noise=.05)
X1, y1 = sklearn.datasets.make_moons(n_samples=1000, noise=.3)
X2, y2 = sklearn.datasets.make_moons(n_samples=200, noise=.05)
X3, y3 = sklearn.datasets.make_moons(n_samples=200, noise=.3)

````

````python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(2,2)
ax[0,0].scatter(X[:,0],X[:,1],c=y)
ax[0,0].axis('equal')

ax[0,1].scatter(X1[:,0],X1[:,1],c=y1)
ax[0,1].axis('equal')

ax[1,0].scatter(X2[:,0],X2[:,1],c=y2)
ax[1,0].axis('equal')

ax[1,1].scatter(X3[:,0],X3[:,1],c=y3)
ax[1,1].axis('equal')
plt.show()
````

# Zad.
Naucz perceptron na danych iris i wylicz F1 score.

````python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

iris = load_iris()
X = iris.data[:, (2, 3)] 
y = (iris.target == 0).astype(np.int32) 

per_clf = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X, y)

y_pred = per_clf.predict(X)

f1 = f1_score(y, y_pred)
print(f1)
plot_decision_regions(X, y, clf=per_clf, zoom_factor=0.2)


#chatgpt solution 

# Wczytaj dane i przekszta je do klasyfikacji binarnej
iris = load_iris()
X = iris.data[:, (2, 3)]  # tylko cechy: patki
y = (iris.target == 0).astype(np.int32)  # 1: setosa, 0: inne

# Podzia na zbir treningowy i testowy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Stwrz i naucz perceptron
clf = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
clf.fit(X_train, y_train)

# Predykcja
y_pred = clf.predict(X_test)

# Oblicz F1 score
f1 = f1_score(y_test, y_pred)

print(f"F1 score: {f1:.4f}")
````
